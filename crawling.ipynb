```python
# CRAWLING DATA PELATIHAN

from bs4 import BeautifulSoup
import urllib.request
import re
import pandas as pd
    
print("Crawl starting ...")

url=[]
content=['','','','','','']
contents=[]
filmList = []
counter=1
df = pd.DataFrame(columns=['deskripsi','luas_tanah(m2)', 'luas_bangunan(m2)', 'jumlah_kamar', 'jumlah_WC', 'harga'])

for x in range(0,63):
    
    print("\n---------------------------------------")
    print("Get Home ",(x+1),"start")
    print("---------------------------------------")

    url="https://rumahdijual.com/carirumah.php?transaksi=BELI&jenis=RUMAH&kota=Bali&minprice=0&maxprice=0&ltmin=0&ktmin=0&q=denpasar&p=%d" % (x+1)
    hdr = { 'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1; rv:57.0) Gecko/20100101 Firefox/57.0' }
    req = urllib.request.Request(url, headers=hdr)
    page = urllib.request.urlopen(req)
    soup = BeautifulSoup(page, 'html.parser')
    regex = re.compile('tblSearchResultRow')

    content_lis = soup.find_all('table', attrs={'class': regex})
    for el in content_lis:
        childTitle = el.findChildren("td", {'class': 'tdInfoSpec hide-sm'}, recursive=True)[0]
        if (len(childTitle) == 4):
            childTitleHarga = childTitle.findChildren("div", recursive=True)[0].getText()
            childTitleTanah = childTitle.findChildren("div", recursive=True)[1].getText()
            childTitleBangunan = childTitle.findChildren("div", recursive=True)[2].getText()
            childTitleRuang = childTitle.findChildren("div", recursive=True)[3].getText()
            
            childTitleHargaSplit = childTitleHarga.split()
            childTitleTanahSplit = childTitleTanah.split()
            childTitleBangunanSplit = childTitleBangunan.split()
            childTitleRuangSplit = childTitleRuang.split()
            
            if(childTitleHargaSplit[1] == 'juta'):
                hargaRumah = float(childTitleHargaSplit[0]) * 1000000
            elif(childTitleHargaSplit[1] == 'miliar'):
                hargaRumah = float(childTitleHargaSplit[0]) * 1000000000
            else:
                hargaRumah = 0
            
        childTitle = el.findChildren("td", {'class': 'TdTitleDesc'}, recursive=True)[0]
        childJudul = childTitle.findChildren("div", recursive=True)[0].getText()
        
        if(len(childTitleRuangSplit) == 2):
            df = df.append({'deskripsi': childJudul, 
                       'luas_tanah(m2)': childTitleTanahSplit[0], 
                       'luas_bangunan(m2)': childTitleBangunanSplit[0], 
                       'jumlah_kamar': childTitleRuangSplit[0],
                       'jumlah_WC': childTitleRuangSplit[1],
                       'harga': hargaRumah}, ignore_index=True)

print("\n---------------------------------------")
print("Crawl finished ...")
df.to_csv('data/rumah_data.csv',  index=False)
```
